```python

import pandas as pd

Europewaste = pd.read_excel("Wasteeuropedatas.xlsx")  # Replace with actual file name if different
Europewaste.head(10)
```

```python
# Assuming your DataFrames are named 'Europewaste2022' and 'EducationEU2022'
Europewaste2022_countries = set(Europewaste2022['Country'])
EducationEU2022_countries = set(EducationEU2022['country'])

# Find the difference (countries in Europewaste2022 but not in EducationEU2022)
difference = Europewaste2022_countries - EducationEU2022_countries

# Print the difference
print("Countries in Europewaste2022 but not in EducationEU2022:", difference)

# Find the difference (countries in EducationEU2022 but not in Europewaste2022)
difference2 = EducationEU2022_countries - Europewaste2022_countries

# Print the difference
print("Countries in EducationEU2022 but not in Europewaste2022:", difference2)
```

```python
Europewaste2022 = Europewaste2022.rename(columns={"Country": "country"})

# Display the updated column names
print(Europewaste2022.columns)
```

```python
print(EducationEU2022)
print(Europewaste2022)
```

```python
# Assuming your DataFrames are named 'EducationEU2022' and 'Europewaste2022'
merged_df = pd.merge(EducationEU2022, Europewaste2022, on='country', how='inner')

# Display the merged DataFrame
print(merged_df)
```

```python
merged_df = merged_df.rename(columns={"2022": "Waste in 2022"})

# Display the updated column names
print(merged_df.columns)
```

```python
import plotly.express as px

fig = px.scatter(merged_df,
                 x="EducationIndex_EducationIndex_2022",
                 y="Waste in 2022",
                 hover_name="country",  # Display country name on hover
                 title="Education Index vs. Waste in 2022")
fig.show()
```
