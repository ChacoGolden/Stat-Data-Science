```python
# PACKAGES
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
url = 'GDP per capita.xls'
GDP = pd.read_excel(url)
GDP.head(10)
```

```python
GDP.columns = ['Country', '2004', '2006', '2008', '2010', '2012', '2014', '2016', '2018', '2020', '2022']
```

```python
print(GDP.columns)
print(GDP.head(2))
```

```python
GDP = GDP.rename(columns={GDP.columns[0]: "Country"})
GDP.head(2)
```

```python
columns_to_average = ['2004', '2006', '2008', '2010', '2012', '2014', '2016', '2018', '2020', '2022']

# Calculate the average and add a new column
GDP['Average'] = GDP[columns_to_average].mean(axis=1)
GDP.head(2)
```

```python
GDP.describe()
```

```python
GDPav = GDP[['Country', 'Average']]
GDPav.head(2)
```

```python
url = 'Waste_percapita.xlsx'
Waste = pd.read_excel(url)
Waste.head(10)
```

```python
Waste.describe()
```

```python
gdp_countries = set(GDPav['Country'])  # Assuming 'Country' is the column name for countries in GDP data
waste_countries = set(Waste['Country'])  # Assuming 'Country' is the column name for countries in Waste data

# Find common countries
common_countries = gdp_countries.intersection(waste_countries)

# Print the common countries
print("Common countries:", common_countries)
```

```python
countries_to_keep = ["Estonia", "France", "Cyprus", "Lithuania", "Italy", "Malta", "Bulgaria", "Belgium", "Iceland", "Croatia", "Bosnia and Herzegovina", "Albania", "Slovakia", "Austria", "Poland", "Greece", "Norway", "Netherlands", "Germany", "Romania", "Denmark", "Luxembourg", "Portugal", "Hungary", "Spain", "Ireland", "Latvia", "Switzerland", "Slovenia", "Sweden", "Finland", "Czechia"]
Waste2 = Waste[Waste['Country'].isin(countries_to_keep)]

print(Waste2.head(2))
```

```python
Waste2.describe()
```

```python
countries_to_remove = ["Switzerland", "Albania", "Bosnia and Herzegovina"]
Waste2 = Waste2[~Waste2['Country'].isin(countries_to_remove)]
```

```python
na_rows = Waste2[Waste2.isna().any(axis=1)]
print(na_rows)
```

```python
Waste2.columns = ['Country', '2004', '2006', '2008', '2012', '2014', '2016', '2018', '2020', '2022']
```

```python
Waste2.tail(2)
```

```python
columns_to_average2 = ['2004', '2006', '2008', '2012', '2014', '2016', '2018', '2020', '2022']

# Calculate the average and add a new column
Waste2['Average'] = Waste2[columns_to_average2].mean(axis=1)
Waste2.head(5)
```

```python
Wasteav = Waste2[['Country', 'Average']]
```

```python
Wasteav = Wasteav.rename(columns={'Average': 'Waste per capita in tonnes'})
print(Wasteav)
```

```python
Wasteav.describe()
```

```python
GDPav.describe()
```

```python
GDPav = GDPav.rename(columns={'Average': 'GDP per capita'})
print(GDPav)
```

```python
# Assuming your dataframes are named 'EducationWW' and 'Europewaste'
merged_df = pd.merge(GDPav, Wasteav, on='Country', how='inner')
print(merged_df)
```

```python
import plotly.express as px

# Assuming your merged dataframe is named 'merged_df'
fig = px.scatter(merged_df,
                 x='Waste per capita in tonnes',  # Assuming 'Waste per capita' is the column name for GDP per capita
                 y='GDP per capita',  # Assuming '2022' is the column name for Waste per capita in tonnes
                 hover_name='Country')  # Display country name on hover

fig.update_layout(
    xaxis_title="Waste per capita in tonnes",
    yaxis_title="GDP per capita",
    title="Waste per capita vs. GDP per capita"
)
```
