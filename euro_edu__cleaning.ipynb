```python
import pandas as pd
import numpy as np
import io
import requests
url_euro_edu = 'https://raw.githubusercontent.com/gaojing33/Stat-Data-Science/refs/heads/main/euro_edu.csv'
```

```python
euro_edu = pd.read_csv(url_euro_edu)
euro_edu.head(10)
```

```python

# Replace ':' with NaN
euro_edu.replace(':', np.nan, inplace=True)
# Check for missing values across the entire dataset
euro_edu.isnull().sum()

# Filter rows (countries) with any missing values
missing_values = euro_edu[euro_edu.isnull().any(axis=1)]

# Display the countries with missing values
missing_values
```

```python
# Drop rows with any missing values
euro_edu_cleaned = euro_edu.dropna()

# Display the cleaned dataset
euro_edu_cleaned
```

```python
# Exclude the 'Country' column from the calculation
# Convert the columns to numeric in case some values are still strings
euro_edu_cleaned.iloc[:, 1:] = euro_edu_cleaned.iloc[:, 1:].apply(pd.to_numeric)

# Calculate the average for each country across all years
euro_edu_cleaned['Average_Education_Level'] = euro_edu_cleaned.iloc[:, 1:].mean(axis=1)

# Create a new dataframe with just the country and the average education level
average_education_df = euro_edu_cleaned[['Country', 'Average_Education_Level']]

# Display the new dataframe
average_education_df
```

```python
# Calculate the median of the average education levels
median_average = average_education_df['Average_Education_Level'].median()

# Display the result
print("The median of the average education levels is:", median_average)
```
